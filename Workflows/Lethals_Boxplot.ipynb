```python
import seaborn
from matplotlib import rcParams
#rcParams.update({'figure.autolayout': True})
import matplotlib.pyplot as plt
import pandas as pd
```

```python
rescue_counts = pd.read_csv('/Users/ashnanarain/Computational_Biology/Gut_Microbiome/Data/FINAL_Species_Wise_Rescue_Counts.csv')
rescue_counts
```

```python
orgs_file = open('/Users/ashnanarain/Computational_Biology/Gut_Microbiome/Lethals/Single_Species.csv', 'r')
orgs_read = orgs_file.readlines()
orgs = []
for i in orgs_read:
    orgs.append(i.strip())
orgs
```

```python
final_counts = pd.DataFrame(columns = orgs, index = range(19))
final_counts
```

```python
for org in final_counts.columns:
    counts_1 = pd.Series(rescue_counts['Count_1'][rescue_counts['Organism_1'] == org])
    counts_2 = pd.Series(rescue_counts['Count_2'][rescue_counts['Organism_2'] == org])
    final_counts[org] = pd.concat([counts_1, counts_2], ignore_index=True)
    
print(final_counts)
```

```python
index_sort = final_counts.mean().sort_values().index
index_sort

final_counts_sorted = final_counts[index_sort]
```

```python
seaborn.boxplot(final_counts_sorted, orient = 'y')
plt.savefig("Lethals_Boxplot.svg")
```
